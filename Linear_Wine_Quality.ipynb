```python
##Clearly regression should be a good choice.
##This is a convinient Data Set 

##Lets start by a simple multiple regression model and latter we shall built upon it#################################
```

```python
x = data.drop("quality",1).as_matrix()
y = data["quality"].as_matrix()
lr = LinearRegression()
lr.fit(x, y)
pred = lr.predict(x)
training_rmse = (((pred - y)**2).mean())**0.5
print (training_rmse)
```

```python
lr.coef_
sns.regplot(y,lr.predict(x))
```

```python
residuals = y-lr.predict(x)
sns.distplot(residuals)
```

```python
##Normality(This dataset is Awesom)
sns.regplot(y,residuals)
```

```python
## Homoscadastictiy(almost) 
plt.plot(residuals)
```

```python
##from the above plot we verify the presence of Independence, absence of any time series relation,

##The particular model staisfies all the assumptions of the Linear regression model

##Validation Via Cross validation
```

```python
rmse_list = []
k = 10 ##folds of cross validation
for K in range(k):
    x_tr = [x[i:i+1] for i in range(len(x)) if i % k != K]
    x_train = np.reshape(x_tr,(len(x_tr),11))
    
    y_tr = [y[i:i+1] for i in range(len(y)) if i % k != K]
    y_train = np.reshape(y_tr,(len(y_tr),1))
    
    x_te = [x[i:i+1] for i in range(len(x)) if i % k == K]
    x_test = np.reshape(x_te,(len(x_te),11))
    
    y_te = [y[i:i+1] for i in range(len(y)) if i % k == K]
    y_test = np.reshape(y_te,(len(y_te),1))
    
    lr = LinearRegression()
    lr.fit(x_train, y_train) 
    
    pred = lr.predict(x_test)
    
    rmse = (((pred - y_test)**2).mean())**0.5
    rmse_list.append(rmse)
    



```

```python
rmse_list
```

```python
CV_RMSE = sum(rmse_list) / float(len(rmse_list))
print(CV_RMSE)
```
